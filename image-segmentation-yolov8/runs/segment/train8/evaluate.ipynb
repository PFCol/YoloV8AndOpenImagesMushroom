```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV data
df = pd.read_csv('results.csv')

# 1. Loss Evolution Over Epochs
fig1, ax1 = plt.subplots(figsize=(15, 5))
ax1.plot(df['                  epoch'], df['         train/box_loss'], label='Train Box Loss', color='blue')
ax1.plot(df['                  epoch'], df['         train/seg_loss'], label='Train Segmentation Loss', color='green')
ax1.plot(df['                  epoch'], df['         train/cls_loss'], label='Train Classification Loss', color='red')
ax1.plot(df['                  epoch'], df['         train/dfl_loss'], label='Train Other Loss', color='purple')
ax1.plot(df['                  epoch'], df['           val/box_loss'], '--', label='Validation Box Loss', color='blue')
ax1.plot(df['                  epoch'], df['           val/seg_loss'], '--', label='Validation Segmentation Loss', color='green')
ax1.plot(df['                  epoch'], df['           val/cls_loss'], '--', label='Validation Classification Loss', color='red')
ax1.plot(df['                  epoch'], df['           val/dfl_loss'], '--', label='Validation Other Loss', color='purple')
ax1.set_title('Loss Evolution Over Epochs')
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Loss Value')
ax1.legend()
plt.show()

# 2. Metrics Evolution Over Epochs
fig2, ax2 = plt.subplots(figsize=(15, 5))
ax2.plot(df['                  epoch'], df['   metrics/precision(B)'], label='Precision (B)', color='blue')
ax2.plot(df['                  epoch'], df['      metrics/recall(B)'], label='Recall (B)', color='green')
ax2.plot(df['                  epoch'], df['   metrics/precision(M)'], '--', label='Precision (M)', color='blue')
ax2.plot(df['                  epoch'], df['      metrics/recall(M)'], '--', label='Recall (M)', color='green')
ax2.set_title('Precision and Recall Evolution Over Epochs')
ax2.set_xlabel('Epochs')
ax2.set_ylabel('Metric Value')
ax2.legend()
plt.show()

# 3. mAP Evolution Over Epochs
fig3, ax3 = plt.subplots(figsize=(15, 5))
ax3.plot(df['                  epoch'], df['       metrics/mAP50(B)'], label='mAP50 (B)', color='blue')
ax3.plot(df['                  epoch'], df['    metrics/mAP50-95(B)'], label='mAP50-95 (B)', color='green')
ax3.plot(df['                  epoch'], df['       metrics/mAP50(M)'], '--', label='mAP50 (M)', color='blue')
ax3.plot(df['                  epoch'], df['    metrics/mAP50-95(M)'], '--', label='mAP50-95 (M)', color='green')
ax3.set_title('mAP Evolution Over Epochs')
ax3.set_xlabel('Epochs')
ax3.set_ylabel('Metric Value')
ax3.legend()
plt.show()

# 4. Learning Rate Evolution Over Epochs
fig4, ax4 = plt.subplots(figsize=(15, 5))
ax4.plot(df['                  epoch'], df['                 lr/pg0'], label='Learning Rate (pg0)', color='blue')
ax4.plot(df['                  epoch'], df['                 lr/pg1'], label='Learning Rate (pg1)', color='green')
ax4.plot(df['                  epoch'], df['                 lr/pg2'], label='Learning Rate (pg2)', color='red')
ax4.set_title('Learning Rate Evolution Over Epochs')
ax4.set_xlabel('Epochs')
ax4.set_ylabel('Learning Rate')
ax4.legend()
plt.show()
```
